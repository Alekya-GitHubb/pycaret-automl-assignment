```python
%%writefile task5_clustering.py
from pycaret.clustering import *
import pandas as pd
import matplotlib.pyplot as plt

# 1️⃣ Load dataset
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
print("✅ Dataset loaded with shape:", df.shape)

# 2️⃣ Set up PyCaret clustering
s = setup(
    data=df,
    normalize=True,
    ignore_features=["species"],  # we only cluster numeric features
    session_id=123,
    use_gpu=True
)

# 3️⃣ Train and compare cluster models
best_model = create_model("kmeans")
print("✅ Model trained successfully!")

# 4️⃣ Assign cluster labels to the dataset
clustered_df = assign_model(best_model)
print("✅ Cluster labels assigned successfully!")
print(clustered_df.head())

# 5️⃣ Visualize clusters
plot_model(best_model, plot="cluster", save=True)
print("✅ Cluster visualization saved!")

# 6️⃣ Save the trained model
save_model(best_model, "/content/iris_cluster_model")
print("✅ Model saved as iris_cluster_model.pkl")
```

```python
%%writefile task6_anomaly_detection.py
from pycaret.anomaly import *
import pandas as pd
import matplotlib.pyplot as plt

# 1️⃣ Load dataset (✅ fixed correct lowercase URL)
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
print("✅ Dataset loaded with shape:", df.shape)

# 2️⃣ Setup PyCaret anomaly detection
s = setup(
    data=df,
    normalize=True,
    ignore_features=["species"],  # non-numeric column
    session_id=123,
    use_gpu=True
)

# 3️⃣ Train Isolation Forest model
model = create_model("iforest")
print("✅ Isolation Forest model created!")

# 4️⃣ Assign anomaly scores
anomaly_df = assign_model(model)
print("✅ Anomaly scores assigned:")
print(anomaly_df.head())

# 5️⃣ Create a custom visualization (since PyCaret 3.x removed anomaly plots)
plt.figure(figsize=(8,5))
plt.hist(anomaly_df["Anomaly_Score"], bins=20, color="lightcoral", edgecolor="black")
plt.title("Distribution of Anomaly Scores (Isolation Forest)")
plt.xlabel("Anomaly Score")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.savefig("/content/Anomaly_Distribution.png")
print("✅ Custom anomaly distribution plot saved as Anomaly_Distribution.png")

# 6️⃣ Save the trained model
save_model(model, "/content/iris_anomaly_model")
print("✅ Anomaly-detection model saved as iris_anomaly_model.pkl")
```
